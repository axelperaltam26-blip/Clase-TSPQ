# **Población por año**

```python
ruta_poblacion="https://raw.githubusercontent.com/marinlopabi/Temas-Selectos-PQ-AD-Python/refs/heads/main/Datos/poblacion.csv"
```

```python
poblacion= pd.read_csv(ruta_poblacion) #siempre usare el comando read para que lea la liga la programaci´ón
```

```python
poblacion.head ()
```

```python
poblacion["1990"] = poblacion["H_1990"] + poblacion["M_1990"]
poblacion["2000"] = poblacion["H_2000"] + poblacion["M_2000"]
poblacion["2010"] = poblacion["H_2010"] + poblacion["M_2010"]
poblacion["2020"] = poblacion["H_2020"] + poblacion["M_2020"]
```

```python
poblacion = poblacion.drop (["H_1990","M_1990","H_2000",
                             "M_2000","H_2010","M_2010",
                             "H_2020","M_2020","Regiones"],axis=1)
```

```python
poblacion = poblacion.set_index("Estado") #siempre usar index para comprimir el indice
poblacion.head()
```

```python
F = ["1990","2000","2010","2020"]
C = [poblacion["1990"].sum(),poblacion["2000"].sum(),
    poblacion["2010"].sum(),poblacion["2020"].sum()]
```

```python
pob = pd.DataFrame ({"Poblacion":C}, index=F)
```

```python
pob #abreviatura de población nos ayuda a acortar los comandos
```

```python
import matplotlib.pyplot as plt
plt.bar(pob.index, pob["Poblacion"], color = "blue", alpha=0.6)
plt.title("Poblacion por año")
plt.ylabel("Poblacion (100,000,000)")
plt.xlabel("Año")
plt.show()
```

# **Clase 2 - Semana 6**

```python
ruta="https://raw.githubusercontent.com/marinlopabi/Temas-Selectos-PQ-AD-Python/refs/heads/main/Datos/CEMEXCPO.csv"
```

```python
df = pd.read_csv(ruta)
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date']) #datetime comando de tiempo
```

```python
df.info()
```

```python
df = df.set_index("Date") #lo podemos usar para datos como fechas, años todo lo relacionado con los tiempos númericos
df
```

```python
df.sum() #calcular todos los valores de cada pilar
```

```python
df.idxmax() #nos sirve para indicar cual el valor más alto de cada indice
```

```python
df.idxmin() #y este para el indicar el valor más bajo
```

```python
df.loc[df.idxmax()[3]][3]
```

```python
df.loc[df.idxmin()[3]][3]
```

```python
plt.figure(figsize=(20,10), dpi=150)
plt.plot(df.index,df["Close"],lw=2, color="black")
plt.xlabel("Fecha", fontdict = {"fontsize":20 })
plt.ylabel("Precio ($)",fontdict = {"fontsize":20 })
plt.title("Precio de Cierre (CEMEX)", fontsize = 25)
plt.tick_params(labelsize=20)
plt.grid(True)

print('Precio de cierre más bajo: ',df.loc[df.idxmin()[3]][3], ', fecha: ',df.idxmin()[3])
print('Precio de cierre más alto: ',df.loc[df.idxmax()[3]][3], ', fecha: ',df.idxmax()[3])
```

```python
plt.figure(figsize=(16, 8), dpi=150)

plt.plot(df.index, df["Close"], lw=2.5, color="black", label="Precio de cierre") #Título

min_idx = df["Close"].idxmin()
max_idx = df["Close"].idxmax()
min_val = df.loc[min_idx, "Close"]
max_val = df.loc[max_idx, "Close"]

plt.scatter(min_idx, min_val, color="red", s=80, label=f"Mínimo: ${min_val:.2f}")
plt.scatter(max_idx, max_val, color="green", s=80, label=f"Máximo: ${max_val:.2f}")

plt.xlabel("Fecha", fontsize=18, labelpad=10)
plt.ylabel("Precio ($)", fontsize=18, labelpad=10)
plt.title("Precio de Cierre - CEMEX", fontsize=22, fontweight="bold", pad=15) #etiquetas

plt.annotate(f"Mínimo\n${min_val:.2f}",
             xy=(min_idx, min_val),
             xytext=(-100, -40),
             textcoords="offset points",
             arrowprops=dict(facecolor="red", arrowstyle="->"),
             fontsize=14, color="red") #para mostar los valores min y max

plt.annotate(f"Máximo\n${max_val:.2f}",
             xy=(max_idx, max_val),
             xytext=(40, 40),
             textcoords="offset points",
             arrowprops=dict(facecolor="green", arrowstyle="->"),
             fontsize=14, color="green")

#configuración del gráfico
plt.grid(alpha=0.3)
plt.tick_params(labelsize=14)
plt.legend(fontsize=14)
plt.tight_layout()

plt.show()

# Mostrar también los datos por consola
print(f" 📉 Precio de cierre más bajo: ${min_val:.2f}, fecha: {min_idx}")
print(f" 📈 Precio de cierre más alto: ${max_val:.2f}, fecha: {max_idx}")
```
