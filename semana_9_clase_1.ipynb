# **Archivos TXT**

Un archivo txt se puede leer con read_csv() porque ambos son solo texto.

Pandas no se fija en la extensión, sino en cómo están separados los datos, si el txt tiene comas, punto y coma, tabulaciones o espacios, read_csv() sabe interpretarlo como tabla.


```python
ruta= "https://raw.githubusercontent.com/marinlopabi/Temas-Selectos-PQ-AD-Python/refs/heads/main/Datos/lista_uami.txt"
df00= pd.read_csv(ruta, sep="\t") #cada delimitador se debe utilizar dependiendo de los datos
df00.head()
```

```python
ruta = "https://raw.githubusercontent.com/marinlopabi/Temas-Selectos-PQ-AD-Python/refs/heads/main/Datos/vent.txt"
df = pd.read_csv(ruta, sep="\t") #cada delimitador se debe utilizar dependiendo de los datos
df.head()
```

```python
ruta = "https://raw.githubusercontent.com/marinlopabi/Temas-Selectos-PQ-AD-Python/refs/heads/main/Datos/signals.txt"
df0 = pd.read_csv(ruta, delim_whitespace=True) #esta delimitación no necesita comillas para leerse
df0.head()
```
