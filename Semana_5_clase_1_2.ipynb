# Uso de funciones para evaluar el estado termodinámico del agua

```python
def a_K (T, unidad):
  if unidad == "C":
      res= T + 273.15
      return res
  elif unidad == "F":
      res= (T - 32) * 5/9 + 273.15
      return res
  elif unidad == "R":
      res= T*5/9
      return res
  else:
    print("Dato no valido, verifique las unidades")
```

```python
def p_MPa(P, unidad):
    if unidad == "bar":
       res= P*0.1
       return res
    elif unidad == "mmHg":
      res = P / 7501
      return res
    elif unidad == "Psi":
       res = P / 145
       return res
    else:
      print("Dato no válido, verifique las unidades")
```

```python
p_MPa(10, "Psi")
```

```python
def T_kelvin(T, unidad):
    if unidad == "C":
        return T + 273.15
    elif unidad == "F":
        return (T - 32)* 5/9 + 273.15
    elif unidad == "R":
        return T* 5/9
    elif unidad == "K":
        return T
    else:
       raise ValueError("Unidad de temperatura no válida. Usa C, F, R o K.")
```

```python
print(T_kelvin(25, "R"))
```

Conversiones P

```python
def P_MPa(P, unidad):
    if unidad == "bar":
        return P * 0.1
    elif unidad == "mmHg":
        return P/7501
    elif unidad == "Psi":
        return P / 145
    elif unidad == "atm":
        return P * 0.101325
    elif unidad == "MPa":
        return P
    else:
       raise ValueError("Unidad de Presión no válida. Usa bar, mmHg, Psi, atm o MPa")
```

```python
print(P_MPa(10, "mmHg"))
```

```python
def clasificar_estado(T_K, P_MPa):
    Tc = 647.1
    Pc = 22.06

    if T_K < Tc and P_MPa < Pc:
        return "Líquido o vapor"
    elif T_K > Tc and P_MPa > Pc:
        return "Supercrítico"
    elif T_K < Tc and P_MPa > Pc:
        return "Líquido comprimido"
    else:
        return "Cerca del punto crítico"
```

```python
clasificar_estado (250, 100)
```

```python
T = float(input("Temperatura: "))
uT = input("Unidad (C, F, R o K): ")

P = float(input("Presión: "))
uP = input("Unidad (bar, mmHg, Psi, atm o MPa): ")

T_K = T_kelvin(T, uT)
P_MPa = P_MPa(P, uP)

print(clasificar_estado(T_K, P_MPa))
```

# **Expresión Regular**

Una expresión regular es una secuencia de caracteres que define un patrón de búsqueda en cadenas de texto y se usa para buscar, validar, reemplazar o dividir texto de forma eficiente y flexible.

```python
capitales = ['#Ciudad Victo*ria', 'cuerna#vaca', 'Morelia!', 'xalapa', 'Tepic', 'hermos*illo',
 'toluca?', 'San? Luis Potosí', '*Mexicali', 'Villahermosa', 'Tuxtla Gutiérrez#',
 'Colima', 'la Paz#', 'Guadalajara', 'C*ampeche!', 'Durango', 'Zacatecas',
 '!Chetumal', 'Ciudad# de México', 'chilpancingo*', 'Chihua#hua', '?Aguascalientes',
 'Oaxaca de Juárez*', '?Puebla', 'Quer?étaro', 'Mérida', 'guanajuato',
 'pachu*ca', '#Saltillo#', 'Culiacán!', 'Tlaxcala', 'monterrey']
```

```python
import re #es la simplificación de la expresión regex

def limpiar_cadenas (strings):
  result = []
  for value in strings:
    value = value.strip()
    value = re.sub("[#*?!]", "", value)
    value = value.title()
    result.append (value)
  return result
```

```python
limpiar_cadenas (capitales) #entre parentesis siempre va la orden a realizar
```

```python
capitales_limpias = limpiar_cadenas (capitales)

capitales_limpias.sort() #sort para ordenar en orden alfabético
```

```python
capitales_limpias
```

Para encontrar X: reemplaza la primera columna de A con el vector de términos independientes.
Para encontrar Y: reemplaza la segunda columna
Para encontrar Z: reemplaza la tercera columna (y así sucesivamente)

```python
X = np.array([[18,4,6],
               [24, 5,6],
               [4, 1, -2]])
```

Calcula el determinante de cada matriz modificada
Dx, Dy, Dz,

```python
Dx = np.linalg.det(X)
Dx
```

Aplica la fórmula de Cramer

1.   x = Dx/D
2.   y = Dy/D
3.   z = Dz/D

```python
x = Dx/D
x
```
