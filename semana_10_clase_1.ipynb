```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ruta="https://raw.githubusercontent.com/marinlopabi/Temas-Selectos-PQ-AD-Python/refs/heads/main/Datos/COVID19MEXICO.csv"
df = pd.read_csv(ruta)
df
```

```python
df.nunique().sort_values(ascending=False) #este comando cuenta cuántos valores unicos hay en cada columna
```

```python
sector_map = {
    1: "CRUZ ROJA",
    2: "DIF",
    3: "ESTATAL",
    4: "IMSS",
    5: "IMSS-BIENESTAR",
    6: "ISSSTE",
    7: "MUNICIPAL",
    8: "PEMEX",
    9: "PRIVADA",
    10: "SEDENA",
    11: "SEMAR",
    12: "SSA",
    13: "UNIVERSITARIO",
    14: "CIJ",
    15: "IMSS Bienestar OPD",
    99: "NO ESPECIFICADO"
}
```

```python
df["SECTOR"] = df["SECTOR"].map(sector_map)
```

```python
df["SECTOR"] #describe la columna indicada entre comillas porqué es texto
```

```python
df["SEXO"] = df["SEXO"].map({1: "MUJER", 2: "HOMBRE",3:"NO ESPECIFICADO"})
```

```python
df["TIPO_PACIENTE"] = df["TIPO_PACIENTE"].map({1:"AMBULATORIO", 2:"HOSPITALIZADO",3:"NO ESPECIFICADO"})
#en el comando siempre escirbir el texto tal cual viene en el archivo o no lo recnocerá
```

```python
clasificacion_covid_map = {
    1: "CONFIRMADO",
    2: "CONFIRMADO",
    3: "CONFIRMADO",
    4: "NO VÁLIDO",
    5: "NO VÁLIDO",
    6: "SOSPECHOSO",
    7: "NEGATIVO"
}
```

```python
df["CLASIFICACION_FINAL_COVID"] = df["CLASIFICACION_FINAL_COVID"].map(clasificacion_covid_map)
```

```python
df["CLASIFICACION_FINAL_COVID"]
```

```python
fechas = ['FECHA_SINTOMAS', 'FECHA_INGRESO','FECHA_DEF', 'FECHA_ACTUALIZACION']
for f in fechas:
    df[f] = pd.to_datetime(df[f], errors='coerce')
    #con este comando transformamos las fechas a datos de tiempo
```

```python
df["FECHA_DEF"]
```

```python
columnas_eliminar = ["ORIGEN", "ENTIDAD_RES", "ID_REGISTRO", "ENTIDAD_UM", "FECHA_DEF", "OTRO_CASO"]
#con este comando podemos eliminar las columnas que no se van a ocupar

df = df.drop(columnas_eliminar, axis=1, errors="ignore") #errors=ignore para evitar si alguna columna mencionada no existe
df
```

```python
df["SEXO"].value_counts() #comando para los conteos númericos
```

```python
conteo_sexo = df["SEXO"].value_counts()
plt.figure(figsize=(6,4))

conteo_sexo.plot(kind="bar", color=["orange", "pink"]) #comando para el tipo de gráfica vertical

plt.title("Número de pacientes por sexo")

plt.xlabel("Sexo") #x para horizontal

plt.ylabel('Cantidad de ingresos') #y para vertical

plt.xticks(rotation=0)

plt.grid(axis="y", linestyle='--', alpha=0.5)

plt.show()
```

```python
plt.figure(figsize=(6,6))

conteo_sexo.plot(kind="pie", autopct="%1.1f%%", startangle=90, colors=["orange", "pink"]) #kind_pie para gráfica circular

plt.ylabel("")

plt.title("Proporción de ingresos por sexo")

plt.show()
```

```python
df_positivos = df[df["CLASIFICACION_FINAL_COVID"] == "CONFIRMADO"] #solo datos confirmados de COVID-19

df_positivos.groupby(["FECHA_INGRESO", "SEXO"]).size().unstack().plot(figsize=(10,5)) #groupby es el comando para agrupar

plt.title("Ingresos diarios por sexo")

plt.ylabel("Número de casos")

plt.show()
```

```python
df_conf = df[df["CLASIFICACION_FINAL_COVID"] == "CONFIRMADO"] #solo datos confirmados

plt.figure(figsize=(8,5))
plt.hist(df_conf["EDAD"], bins=30, color="blue", edgecolor='black', alpha=0.7)

plt.title("Distribución de edad en casos confirmados de COVID-19")

plt.xlabel("Edad")

plt.ylabel("Número de casos")

plt.grid(axis="y", linestyle="--", alpha=0.4)

plt.show()
```

```python
df_conf["EDAD"].value_counts().sort_index().plot(
    kind="bar", figsize=(14,5), color="darkcyan"
) #value_counts, cuenta solo los casos por edad e index para ordenar el índice

plt.title("Casos confirmados por edad")
plt.xlabel("Edad")
plt.ylabel("Número de casos")

plt.show()
```

```python
bins = [0, 17, 29, 49, 64, 120]

labels = ["0–17", "18–29", "30–49", "50–64", "65+"]

df_conf["grupo_edad"] = pd.cut(df_conf["EDAD"], bins=bins, labels=labels, right=True) #cut divide los datos númericos


df_conf["grupo_edad"].value_counts().sort_index().plot(
    kind="bar", color="salmon", figsize=(7,5)
)

plt.title("Casos confirmados por grupo de edad")

plt.xlabel("Grupo de edad")

plt.ylabel("Número de casos")


plt.show()
```

```python
df_conf = df[df["CLASIFICACION_FINAL_COVID"] == "CONFIRMADO"] #muestra el producto final


plt.figure(figsize=(8,5))

plt.hist(df_conf["EDAD"], bins=30, color="steelblue", edgecolor="black", alpha=0.7)


plt.title("Distribución de edad en casos confirmados de COVID-19")

plt.xlabel("Edad")

plt.ylabel("Número de casos")

plt.grid(axis="y", linestyle="--", alpha=0.4)

plt.show()
```
