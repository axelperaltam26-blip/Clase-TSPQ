# Control de Flujo

Python tiene varias palabras clave internas para lógica condicional, bucles y otros conceptos estándares de control de flujo, que pueden encontrarse en otros lenguajes de programación.

**if, elif, else**

if → se ejecuta si la condición es verdadera.

elif → se ejecuta si la anterior fue falsa, pero esta verdadera.

else → se ejecuta si ninguna condición anterior fue verdadera. No lleva condición

```python
x = -56
if x <0:
  print("el numero es negativo")
elif x == 0:
  print("el numero es cero")
else:
  print("el numero es positivo")
```

# **Clase 2 - Semana 3**

1. Ley de los gases ideales

```python
n = 2.0          # mol
R = 0.0821       # L·atm/mol·K
T = 298          # K
V = 10           # L

P = (n * R * T) / V
print(f"Presión del gas: {P:.2f} atm")
```

2. Ecuación de Arrhenius (cinética química)

```python
import math
# Datos
A = 5e7          # factor preexponencial [1/s]
Ea = 75000        # energía de activación [J/mol]
R = 8.314         # constante de los gases [J/mol·K]
T = 350           # temperatura [K]

# Ecuación de Arrhenius
k = A * math.exp(-Ea / (R * T))
print(f"Constante de velocidad k = {k:.4e} 1/s")
```

3. Ecuación de Antoine (presión de vapor)

```python
import math
#constantes de antoine para agua (T en *C, P en mmHg)
A, B, C = 8.07131, 1730.63, 233.426
T = 80 # °C
P = 10 ** (A-(B/(C+T)))
print (f"presion de vapor del agua a {T}) °C: {P:.3f} mmHg") #.2f nos dice cuantos decimales queremos en la operacion.
```

4. Cálculo de número de Reynolds (flujo de fluidos)

```python
import math
rho = 1000 # densidad [kg/m3]
V = 1.2    # velocidad [m/s]
D = 0.05   # diametro [m]
mu = 0.001 # viscosidad [Pa.s]
Re =(rho*V*D)/mu
#Re
print (f"El Reynolds es de: {Re:.3f}") #igual las llaves nos ayudan e identificar los valores que queremos encontrar.
```
