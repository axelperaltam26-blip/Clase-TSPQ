# Funciones

Una función es un conjunto de instrucciones que se agrupan para realizar una tarea específica, sirven para reutilizar códigos, organizar mejor los programas y evitar repetir instrucciones.

1. def	palabra clave para definir una función
2. parámetros	datos que la función recibe
3. return	indica qué valor devuelve la función
4. llamada	cuando usamos la función en el programa


```python
def sumar(a, b):  #una operación númerica (siempre la debemos llamar)
    resultado = a + b
    return resultado
```

```python
sumar(4324,234)
```

```python
def resta(a,b):
    res = a - b
    return res
```

```python
resta(19, 27)
```

Crear una función que convierta temperatura Celsius a Kelvin.

```python
def conversionC_K(T_C):
    res= T_C + 273.15
    return res
```

Uso de funciones para evaluar el estado termodinámico del agua

Función que cambie de C a K, F a K

```python
def a_K(T, unidad):
    if unidad == "C":
       res= T +273.15
       return res
    elif unidad == "F":
      res = (T - 32)* 5/9 +273.15
      return res
    elif unidad == "R":
       res = T* 5/9
       return res
    else:
      print("Dato no válido, verifique las unidades")
```

```python
a_K (100, "R")
```

```python
a_K(100, "h")
```
